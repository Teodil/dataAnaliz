## 2. Визуализация средствами Seaborn

В библиотеке есть достаточно сложные типы визуализации, которые в matplotlib потребовали бы большого количество кода.

Познакомимся с первым типом графиков pair plot (scatter plot matrix). Эта визуализация поможет нам посмотреть на одной картинке, как связаны между собой различные признаки

```python
df1.head()
```

```python
sns.pairplot(df1[['Возраст', 'Доход (тыс. руб.)', 'Срок проживания']]);
```

Также с помощью seaborn можно построить распределение

```python
sns.distplot(df1['Срок проживания'])
```

```python
sns.pairplot(df1, hue="Благонадежный заемщик")
```

Для того чтобы подробнее посмотреть на взаимосвязь двух численных признаков, есть еще и joint_plot – это гибрид scatter plot и histogram (отображаются также гистограммы распределений признаков)

```python
sns.jointplot(x='Месячный платеж (тыс.руб.)', y='Доход (тыс. руб.)', 
              data=df1, kind='scatter');
```

```python
sns.boxplot(x="Доход (тыс. руб.)", y="Иждивенцы", data=df1, orient="h");
```

```python
sns.boxplot(y="Доход (тыс. руб.)", x="Благонадежный заемщик", data=df1);
```

```python
sns.barplot(y='Срок проживания',x="Благонадежный заемщик",data=df1)
```

Еще одна полезная визуализация - это **heat map**.

Heat map позволяет посмотреть на распределение какого-то численного признака по двум категориальным.

```python
pivot  = df1.pivot_table(index='Иждивенцы', 
                        columns='Благонадежный заемщик', 
                        values='Доход (тыс. руб.)', 
                        aggfunc=np.mean)
```

```python
pivot
```

```python
sns.heatmap(pivot, annot=True, fmt=".1f", linewidths=.5);
```

## 3. Визуализация средствами Plotly

Plotly - это open-source библиотека, которая позволяет строить интерактивные графики в jupyter.notebook

Перед началом работы импортируем все необходимые модули и инициализируем plotly с помощью команды init_notebook_mode.

```python
!pip install plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
df1.head()
```

```python
dynamic = df1[['Доход (тыс. руб.)', 'Иждивенцы']].groupby('Иждивенцы').mean()
```

```python
dynamic
```

```python
data = [go.Scatter(x=dynamic.index, y=dynamic['Доход (тыс. руб.)'])]
fig = go.Figure(data=data)
iplot(fig)
```

```python
#экспорт графиков в локальный html
plotly.offline.plot(fig, filename='export_income.html', show_link=False);
```

```python
data = [go.Box(x=df1['Иждивенцы'], y=df1['Доход (тыс. руб.)'])]
fig = go.Figure(data=data)
iplot(fig)
```
