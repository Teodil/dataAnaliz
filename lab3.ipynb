```python
from sklearn.feature_extraction.text import TfidfVectorizer
import umap.umap_ as umap
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px

from sklearn.cluster import KMeans

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt
```

```python
file1 = open("/content/drive/MyDrive/ParseData/TestChecked2021_ready.txt", "r")
lines = file1.readlines()
count = 0
texts = [];
#print(len(lines))
print(lines[count+7+7+6])

while len(texts)<1000:
  if count == 0:
    add = 5
  else:
    add = 6
  text = lines[count+add]
  add += 1
  #print(count+add)
  #print('->'+text)
  while '=====' not in lines[count+add]:
    text += ' '+lines[count+add]
    #print('->'+text)
    add +=1
  texts.append(text)
  count += add  

file1.close()
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

from ipywidgets import interact, IntSlider, fixed, FloatSlider

import pymorphy2 # Морфологический анализатор.
import re 
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from matplotlib.cm import autumn
```

```python
import nltk
from nltk.stem.snowball import SnowballStemmer
import re 
  
#the stemmer requires a language parameter
snow_stemmer = SnowballStemmer(language='russian')
  
#list of tokenized words
#tests = ['cared','university','fairly','easily','singing',
#       'sings','sung','singer','sportingly']
  
#stem's of each word
normolizeTexts = []
for i in texts:
  tokens = re.findall('[A-Za-zА-Яа-яЁё]+\-[A-Za-zА-Яа-яЁё]+|[A-Za-zА-Яа-яЁё]+', i)
  normolizeText = ''
  for w in tokens:
      x = snow_stemmer.stem(w)
      normolizeText += x+" "
      #stem_words.append(x)
  normolizeTexts.append(normolizeText)
  
print(normolizeTexts)
```

```python
morph=pymorphy2.MorphAnalyzer() # Создает объект морфоанализатора и загружет словарь.
```

```python
# Pymorphy
def normalizePymorphy(texts):
  normolizeTexts = []
  for text in texts:
    tokens = re.findall('[A-Za-zА-Яа-яЁё]+\-[A-Za-zА-Яа-яЁё]+|[A-Za-zА-Яа-яЁё]+', text)
    sentence = ''
    for t in tokens:
        pv = morph.parse(t)
        #words.append(pv[0].normal_form) # Берем наиболее вероятную форму.
        sentence += pv[0].normal_form+" "
    #return words
    normolizeTexts.append(sentence)
  return normolizeTexts   
        
# Обратите внимание, что про иностранные слова словарь ничего не знает.
normolizeTexts = normalizePymorphy(texts)
```

```python
vectorizer = TfidfVectorizer(token_pattern="[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+-[а-яёА-ЯЁ]+|[а-яёА-ЯЁ]+")
X = vectorizer.fit_transform(normolizeTexts)
tfidf_res = vectorizer.fit_transform(normolizeTexts)
tfIdf = pd.DataFrame(tfidf_res.toarray(),columns=vectorizer.get_feature_names())
```

```python
def sil_demo(k=2):
    
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    
    fig, ax = plt.subplots(1,2)
    
    sil = silhouette_samples(X, kmeans.labels_)
    ax[0].scatter(X[:,0], X[:,1], c=sil)
    
    ax[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                  marker='o', c="white", alpha=1, s=200)
    
    ax[0].set_xlabel('$x_1$')
    ax[0].set_ylabel('$x_2$')

    for i, c in enumerate(kmeans.cluster_centers_):
        ax[0].scatter(c[0], c[1], marker=f'${i}$', alpha=1, s=50)
        
        
        
    ax[1].scatter(k, crit[k-2], color='g', marker='s', s=90)
    ax[1].plot(range(2, max_num), crit, marker='s')
    
    ax[1].set_xlabel('$k$')
    ax[1].set_ylabel('$L^{(k)}(C)$')
    
    plt.show()

    

crit = []

max_num = 12
for k in range(2, max_num):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    crit.append(silhouette_score(X, kmeans.labels_))
```
