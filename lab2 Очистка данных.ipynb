Посмотрим как это работает с DataFrame

Посчитаем количество пропусков

Можно удалять как колонки так и строки. По умолчанию - удаляются строки. Задав параметр axis - можно удалить колонку

```python
df2=df.dropna(axis='columns')
df2
```

Можно управлять режимом удаления и удалять, например, ТОЛЬКО те строки или столбцы, где все элементы - пустые. Это делает how='all'

```python
df['NC']=None
df.head()
```

```python
df=df.dropna(how='all',axis='columns')
df
```

## 1.2 Замена специальными значениями

```python
primer = pd.Series(['1', '2', None, '4', np.nan])
primer
```

Для заполнения пустых значений чаще всего используют метод fillna() - он позволяет как заменить каким-то одним значением, так и обработать по алгоритму какие-то пустые значения

```python
primer = primer.fillna(0)
primer
```

```python
df = pd.read_excel('lab_1_1.xlsx')
df.shape
df
```

Также можно заполнять пропуски в конкретной колонке с использованием своего типа заполняителя для каждой колонки.

```python
df.fillna({'IL_23': 'анализ не взят'})
```

```python
primer = pd.Series([1, 2, np.nan, 4, np.nan])
```

Таже можно заполнять используя функции аггрегации (например, среднее)

```python
primer.fillna(primer.mean())
```

```python
df.fillna({'IL_6': df.IL_16.mean()})
```

 ## 1.3 Замена интерполяцией

Интерполяция – это техника в Python, с которой можно оценить неизвестные точки данных между двумя известными

Линейная интерполяция - заполнение средним между думя соседними значениями признака

```python
primer = pd.Series([1, 2, np.nan, 4, 5,np.nan])
primer
```

```python
primer.interpolate()
```

```python
df
```

```python
df.interpolate()
```

Полиномиальная интерполяция

```python
primer = pd.Series([1,2,3,6,8,np.nan,8,4,5,8,7])
primer
```

```python
primer1=primer.interpolate(method='polynomial', order=5)
primer1
```

```python
primer1.plot()
```

```python
df['New']=df['IL_6'].interpolate(method='polynomial', order=2)
df
```
